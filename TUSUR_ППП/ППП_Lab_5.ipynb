Вывести в одном окне, но на разных графиках (см. subplot), функцию в виде 2 контурных графиков разного типа: contour, contourf.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Исходные данные
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Вычислить функцию
Z = X**2 + Y**2

# Создать фигуру и оси
fig = plt.figure(figsize=(12, 12))

# Трёхмерные графики
ax1 = fig.add_subplot(331, projection='3d')
ax1.plot_wireframe(X, Y, Z, color='b')
ax1.set_title('plot_wireframe')

ax2 = fig.add_subplot(332, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='viridis')
ax2.set_title('plot_surface')

ax3 = fig.add_subplot(333, projection='3d')
ax3.plot_trisurf(X.ravel(), Y.ravel(), Z.ravel(), cmap='viridis')
ax3.set_title('plot_trisurf')

ax4 = fig.add_subplot(334, projection='3d')
ax4.plot(X.ravel(), Y.ravel(), Z.ravel(), 'b-')
ax4.set_title('plot')

ax5 = fig.add_subplot(335, projection='3d')
ax5.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax5.set_title('surfl')

# Контурные графики
ax6 = fig.add_subplot(336)
ax6.contour(X, Y, Z, 10, colors='k')
ax6.set_title('contour')

ax7 = fig.add_subplot(337)
ax7.contourf(X, Y, Z, 10, cmap='viridis')
ax7.set_title('contourf')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# функция правой части дифференциального уравнения
def f(y, x):
    return 0.7*y + x*np.log(x+y)

# начальные условия
y0 = 2.5

# интервал для поиска решения
x = np.arange(12.4, 14.1, 0.08)

# решение дифференциального уравнения с помощью odeint
y = odeint(f, y0, x)

# вывод графика
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Решение дифференциального уравнения')
plt.grid()
plt.show()
```

**Решение методом Эйлера***

```python
import numpy as np
import matplotlib.pyplot as plt

# функция правой части дифференциального уравнения
def f(x, y):
    return 0.7*y + x*np.log(x+y)

# начальные условия
x0 = 12.4
y0 = 2.5

# интервал для поиска решения
x = np.arange(x0, 14.1, 0.08)
n = len(x)

# шаг изменения
h = 0.08

# массив для хранения решения
y = np.zeros(n)
y[0] = y0

# метод Эйлера
for i in range(n-1):
    y[i+1] = y[i] + h*f(x[i], y[i])

# вывод графика
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Решение дифференциального уравнения методом Эйлера')
plt.grid()
plt.show()
```

## Задание 4.

**Модель физической задачи:**

Пусть мы имеем пружинный маятник, состоящий из точки массы m, подвешенной на пружине с коэффициентом упругости k. Маятник колеблется в вертикальной плоскости под действием силы тяжести mg. Пусть x(t) - смещение точки массы от равновесного положения, а y(t) - скорость точки массы.

Тогда система дифференциальных уравнений, описывающая колебания маятника, имеет вид:

dx/dt = y dy/dt = -k/m * x - g

где g - ускорение свободного падения.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# константы
m = 1.0  # масса точки
k = 2.0  # коэффициент упругости
g = 9.81  # ускорение свободного падения

# функция правой части системы уравнений
def f(state, t):
    x, y = state
    dxdt = y
    dydt = -k/m * x - g
    return [dxdt, dydt]

# начальные условия
x0 = 1.0  # начальное смещение
y0 = 0.0  # начальная скорость
state0 = [x0, y0]

# интервал времени
t = np.arange(0, 10, 0.01)

# решение системы уравнений с помощью odeint
state = odeint(f, state0, t)

# вывод графиков
plt.plot(t, state[:, 0], label='x(t)')
plt.plot(t, state[:, 1], label='y(t)')
plt.xlabel('t')
plt.ylabel('x, y')
plt.title('Колебания пружинного маятника')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# константы
m = 1.0  # масса точки
k = 2.0  # коэффициент упругости
g = 9.81  # ускорение свободного падения

# начальные условия
x0 = 1.0  # начальное смещение
y0 = 0.0  # начальная скорость

# интервал времени
t = np.arange(0, 10, 0.01)
n = len(t)

# шаг изменения
h = 0.01

# массивы для хранения решений
x = np.zeros(n)
y = np.zeros(n)
x[0] = x0
y[0] = y0

# метод Эйлера
for i in range(n-1):
    x[i+1] = x[i] + h*y[i]
    y[i+1] = y[i] - h*(k/m * x[i] + g)

# вывод графиков
plt.plot(t, x, label='x(t)')
plt.plot(t, y, label='y(t)')
plt.xlabel('t')
plt.ylabel('x, y')
plt.title('Колебания пружинного маятника методом Эйлера')
plt.legend()
plt.grid()
plt.show()
```

### Оценка результатов:

Оба кода создали графики колебаний системы пружина-масса, которые демонстрируют периодическое движение массы. Кривые x(t) и y(t) имеют синусоидальную форму, как и ожидалось для маятниковой системы.

Ручной метод Эйлера был менее точным, чем реализация scipy, при использовании больших временных шагов. Это может быть связано с тем, что ручной метод Эйлера является приблизительным и не учитывает всех сложностей системы.

В целом, результаты показывают, что как библиотека scipy, так и ручной метод интегрирования по Эйлеру могут быть использованы для решения систем дифференциальных уравнений для модели пружинно-массовой системы с приемлемой точностью.
